Responda las siguientes preguntas (0.2 pts c/u):

1. ¬øQu√© es `Git` y c√≥mo se relaciona con el control de versiones? 

    - `Git` es un Sistema de Control de Versiones Distribuido (DVCS), este administra los registros a trav√©s de archivos en una carpeta .git. Basicamente se registran los cambios realizados a lo largo del tiempo en los archivos de un proyecto, tal y como lo hace una VCS, de ahi viene la relacion con el control de las versiones.

    
2. Nombre y explique 3 diferencias entre `Git` y `Github`.

    - La primera diferencia es por el tipo de uso que tienen, `Git` es un DVCS el cual se encarga de rastrear/gestionar cambios en archivos de un proyecto, por otro lado `Github` es una **plataforma** basada en `Git` que entrega funcionalidades de esta, tal como VCS, pero a√±ade otras caracteristicas, como almacenamiento en la nube, herramientas de colaboraci√≥n, etc.

    - La segunda es la forma de usar que tienen, en este caso `Git` es una herramienta de linea de comando mientras que `Github` se utiliza principalmente a trav√©s de su interfaz web, lo que significa que no es necesario utilizar la l√≠nea de comando para la mayor√≠a de las operaciones.

    - Otra diferencia es que en `Github` se puede subir una copia de los repositorios de `Git`. Tambien podria considerarse que `Git` es un *software* en donde `Github` lo utiliza como base para dar un servicio de alojamiento basado en web.


3. ¬øQu√© es un `hash` y c√≥mo se representa en un `commit`? De un ejemplo.

    - Es una secuencia de caracteres (40) el cual representa de manera unica a un objeto, tales como *blobs*, *commits* o arboles. Como se mencion√≥, el valor es unico debido a que se utilizan para referenciar a un elemento de inter√©s. Por ejemplo en esta linea de codigo:

        <code>commit 35e7f8c9ae4f71a4e6bdf8d1af9d07f6db6ef786</code>

     `35e7f8c9ae4f71a4e6bdf8d1af9d07f6db6ef786` seria el hash unico que representa al `commit`.


4. ¬øQue elementos componen a la estructura de datos de `Git`?

    Los elementos que componen la estructura se detallan a continuacion:
    - **Blob:** Es donde se guarda todo el contenido de un archivo en un momento especifico de la historia del repositorio. Cuando un archivo es modificado, un nuevo blob almacenar√° el archivo completo con todos los nuevos cambios. Asi tambien, cada blob se identifica mediante su hash SHA-1 √∫nico.
    - **Tree:** Representa un directorio y su estructura en un punto espec√≠fico en la historia del repositorio. Contiene referencias a blobs y sub√°rboles, junto con los nombres de los archivos y las rutas que los identifican, siempre habr√° un arbol en la raiz apuntando al arbol que contenga cosas.
    - **Commit:** Representa un punto en la historia del repositorio y contiene informaci√≥n sobre el autor, la fecha, el mensaje de confirmaci√≥n y un puntero al √°rbol que representa el estado de los archivos en ese commit.
    - **Tag**(en internet igual lo consideraban asi que lo agregue igual): Representa una referencia a un commit espec√≠fico. Los tags se utilizan para marcar puntos espec√≠ficos en la historia, como versiones estables del *software*.

    Tambien podria considerar los hash que son parte esencial de la estructura de datos en `Git`, los cuales se utilizan para identificar de manera √∫nica cada objeto almacenado en el repositorio (como *blobs*, *trees* y *commits*).
    

5. ¬øQue comandos de `Git` sirven para crear un repositorio y hacer seguimiento a cambios en los archivos?

    - Para inicializar un repositorio se utiliza: <code> git init nombre-del-repo </code>.
    - Ahora bien, para hacer seguimiento a cambios (en caso de no hacerlos todavia), se deben a√±adir los archivos modificados al *staging area* mediante el comando <code> git add </code>, en caso de a√±adir un solo archivo se debe usar <code> git add path/to/file.py </code>, en caso de ser una carpeta se debe usar <code> git add path/* </code>, y si se quiere agregar todo se debe usar <code> git add --all </code>. Luego se debe realizar el commit que permita capturar una instantanea de los archivos staged del repositorio y agregarlos a su historial de cambios, para ello se ocupa <code> git commit </code>, en caso de querer a√±adir un comentario se debe usar <code> git commit -m "comentario" </code>. Por √∫ltimo, para observar el registro de cambios se utiliza <code> git log </code>, al ejecutar este comando se aprecia la estructura de un commit.

11. `git stash` (0.25 puntos)

12. `git reset`. ¬øCu√°l es la diferencia de este comando con `git revert`?. (0.25 puntos)
    - Se utiliza para mover la rama actual y la referencia del HEAD a un *commit* espec√≠fico o para manipular el estado de la rama y el *staging area*, dependiendo de como se use puede tener diferentes efectos, como <code>-- soft</code>, <code>-- hard</code> o <code>-- mixed</code>, en todos ellos se utiliza, por ejemplo, si se desea deshacer algunos *commits* o reorganizar el historial de *commits*. 
    - La diferencia con `git revert` es que este √∫ltimo crea un nuevo *commit* que deshace los cambios de un *commit* espec√≠fico sin reescribir la historia, a diferencia de `git reset` que cambia el estado de la rama actual y el puntero HEAD, eliminando *commits* del historial.
