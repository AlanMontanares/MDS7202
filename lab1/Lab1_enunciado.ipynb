Responda las siguientes preguntas (0.2 pts c/u):

1. ¿Qué es `Git` y cómo se relaciona con el control de versiones? 

    - `Git` es un Sistema de Control de Versiones Distribuido (DVCS), este administra los registros a través de archivos en una carpeta .git. Basicamente se registran los cambios realizados a lo largo del tiempo en los archivos de un proyecto, tal y como lo hace una VCS, de ahi viene la relacion con el control de las versiones.

    
2. Nombre y explique 3 diferencias entre `Git` y `Github`.

    - La primera diferencia es por el tipo de uso que tienen, `Git` es un DVCS el cual se encarga de rastrear/gestionar cambios en archivos de un proyecto, por otro lado `Github` es una **plataforma** basada en `Git` que entrega funcionalidades de esta, tal como VCS, pero añade otras caracteristicas, como almacenamiento en la nube, herramientas de colaboración, etc.

    - La segunda es la forma de usar que tienen, en este caso `Git` es una herramienta de linea de comando mientras que `Github` se utiliza principalmente a través de su interfaz web, lo que significa que no es necesario utilizar la línea de comando para la mayoría de las operaciones.

    - Otra diferencia es que en `Github` se puede subir una copia de los repositorios de `Git`. Tambien podria considerarse que `Git` es un *software* en donde `Github` lo utiliza como base para dar un servicio de alojamiento basado en web.


3. ¿Qué es un `hash` y cómo se representa en un `commit`? De un ejemplo.

    - Es una secuencia de caracteres (40) el cual representa de manera unica a un objeto, tales como *blobs*, *commits* o arboles. Como se mencionó, el valor es unico debido a que se utilizan para referenciar a un elemento de interés. Por ejemplo en esta linea de codigo:

        <code>commit 35e7f8c9ae4f71a4e6bdf8d1af9d07f6db6ef786</code>

     `35e7f8c9ae4f71a4e6bdf8d1af9d07f6db6ef786` seria el hash unico que representa al `commit`.


4. ¿Que elementos componen a la estructura de datos de `Git`?

    Los elementos que componen la estructura se detallan a continuacion:
    - **Blob:** Es donde se guarda todo el contenido de un archivo en un momento especifico de la historia del repositorio. Cuando un archivo es modificado, un nuevo blob almacenará el archivo completo con todos los nuevos cambios. Asi tambien, cada blob se identifica mediante su hash SHA-1 único.
    - **Tree:** Representa un directorio y su estructura en un punto específico en la historia del repositorio. Contiene referencias a blobs y subárboles, junto con los nombres de los archivos y las rutas que los identifican, siempre habrá un arbol en la raiz apuntando al arbol que contenga cosas.
    - **Commit:** Representa un punto en la historia del repositorio y contiene información sobre el autor, la fecha, el mensaje de confirmación y un puntero al árbol que representa el estado de los archivos en ese commit.
    - **Tag**(en internet igual lo consideraban asi que lo agregue igual): Representa una referencia a un commit específico. Los tags se utilizan para marcar puntos específicos en la historia, como versiones estables del *software*.

    Tambien podria considerar los hash que son parte esencial de la estructura de datos en `Git`, los cuales se utilizan para identificar de manera única cada objeto almacenado en el repositorio (como *blobs*, *trees* y *commits*).
    

5. ¿Que comandos de `Git` sirven para crear un repositorio y hacer seguimiento a cambios en los archivos?

    - Para inicializar un repositorio se utiliza: <code> git init nombre-del-repo </code>.
    - Ahora bien, para hacer seguimiento a cambios (en caso de no hacerlos todavia), se deben añadir los archivos modificados al *staging area* mediante el comando <code> git add </code>, en caso de añadir un solo archivo se debe usar <code> git add path/to/file.py </code>, en caso de ser una carpeta se debe usar <code> git add path/* </code>, y si se quiere agregar todo se debe usar <code> git add --all </code>. Luego se debe realizar el commit que permita capturar una instantanea de los archivos staged del repositorio y agregarlos a su historial de cambios, para ello se ocupa <code> git commit </code>, en caso de querer añadir un comentario se debe usar <code> git commit -m "comentario" </code>. Por último, para observar el registro de cambios se utiliza <code> git log </code>, al ejecutar este comando se aprecia la estructura de un commit.


Responda si las siguientes afirmaciones son Verdaderas (V) o Falsas (F). Justifique en **cualquiera** de los casos (0.2 pts c/u):


6. `Git` siempre resuelve automáticamente los problemas de lineas al hacer merge entre diferentes ramas.
    - **Falso:** Cuando ocurren estas colisiones (llamados conflictos), se debe hacer resolver de forma manual y posteriormente subir los cambios. 
7. Ejecutar `git pull` sobre un repositorio es equivalente a ejecutar `git clone`.
    - **Falso:** `git pull` se utiliza para obtener los últimos cambios de un repositorio remoto y fusionarlos con la rama actual, mientras que `git clone` se usa para clonar un repositorio remoto completo en un nuevo directorio local (*merging*).
8. `git status` genera los mismos resultados que `git log`.
    - **Falso:** `git status` muestra el estado actual de los archivos en tu directorio de trabajo en relación con el repositorio, indicando si hay cambios no confirmados en los archivos, si hay archivos nuevos o eliminados que aún no se han añadido o eliminado del área de preparación, y si el área de preparación está lista para un nuevo commit, mientras que `git log` muestra el historial de *commits* en la rama actual.

10. El comando `git commit` se utiliza para enviar cambios al repositorio remoto.
    - **Falso**: `git commit` se utiliza para crear un nuevo *commit* con los cambios realizados en el repositorio local pero no al repositorio remoto. Para enviar los cambios al repositorio remoto  se debe usar el comando `git push`.

Por último, investigue y comente para qué sirven los siguientes comandos:

11. `git stash` (0.25 puntos)
    - Se utiliza para guardar temporalmente los cambios no comprometidos en un área de almacenamiento temporal llamada "stash", esto permite cambiar de la rama actual a otra y realizar otras acciones sin perder los cambios actuales.
12. `git reset`. ¿Cuál es la diferencia de este comando con `git revert`?. (0.25 puntos)
    - Se utiliza para mover la rama actual y la referencia del HEAD a un *commit* específico o para manipular el estado de la rama y el *staging area*, dependiendo de como se use puede tener diferentes efectos, como <code>-- soft</code>, <code>-- hard</code> o <code>-- mixed</code>, en todos ellos se utiliza, por ejemplo, si se desea deshacer algunos *commits* o reorganizar el historial de *commits*. 
    - La diferencia con `git revert` es que este último crea un nuevo *commit* que deshace los cambios de un *commit* específico sin reescribir la historia, a diferencia de `git reset` que cambia el estado de la rama actual y el puntero HEAD, eliminando *commits* del historial.
