**1. La función `generar_descripcion(fruta)` que reciba una fruta genere una descripción de la fruta de la siguiente manera:**

```{El/La} {...} es una fruta de tipo {...} de color {...}. {Presenta/No presenta} pepas en su interior.```
    
Por ejemplo, `generar_descripcion(plátano)` debe generar el siguiente string:
    
```
'El plátano es una fruta de tipo tropical de color amarillo. No presenta pepas en su interior.'
```
    
Indicaciones: 
- Usen formateo o suma de strings, if, else y for.
- Para saber si ocupar 'El' o 'La', pueden acceder al nombre de la fruta y ver cual es el último carácter a través de indexadores `string[indice]`.
- Utilice los siguientes test para corroborar el funcionamiento de su código:

```python
assert generar_descripcion(naranja) == "La naranja es una fruta de tipo cítrica de color naranjo. Presenta pepas en su interior."
assert generar_descripcion(platano) == "El plátano es una fruta de tipo tropical de color amarillo. No presenta pepas en su interior."
assert generar_descripcion(piña) == "La piña es una fruta de tipo tropical de color amarillo. No presenta pepas en su interior."
assert generar_descripcion(frutilla) == "La frutilla es una fruta de tipo bosque de color rojo. Presenta pepas en su interior."
assert generar_descripcion(frambuesa) == "La frambuesa es una fruta de tipo bosque de color rojo. Presenta pepas en su interior."
```

````python
def generar_descripcion(fruta):
    """
    Genera una descripción de una fruta en base a sus propiedades.

    Esta función toma un diccionario que representa una fruta y genera una descripción
    en base a la informacion del diccionario.

    Parameters:
    -----------
    fruta : dict
        Un diccionario que contiene información sobre la fruta, con las siguientes claves:
        - 'nombre': El nombre de la fruta.
        - 'tipo': El tipo de la fruta.
        - 'color': El color de la fruta.
        - 'pepas': Un valor booleano que indica si la fruta presenta pepas en su interior.

    Returns:
    --------
    str
        Una descripción generada en base a las propiedades de la fruta.

    """
    #Nos aseguramos que el diccionario tenga los valores de las llaves en el formato correcto
    if isinstance(fruta["nombre"], str) and isinstance(fruta["tipo"], str) and isinstance(fruta["color"], str) and isinstance(fruta["pepas"], bool):

        #Principalmente este try-except se usa porque usamos indices, y puede ocurrir que hayan diccionarios con nombre "", por lo que no se puede tratar con indices esos casos
        try:
            if "a" == fruta["nombre"][-1]: #Vemos que la ultima letra sea "a" para asignarle el pronombre "La"
                pronombre = "La"
            else:
                pronombre = "El"
        except: # En caso "" la tratamos como "La" (fruta desconocida) 
            pronombre = "La"   
            fruta["nombre"] = "-fruta desconocida-"
        
        if fruta["tipo"] =="": #Hacemos lo mismo para el tipo
            fruta["tipo"] = "-desconocido-"

        if fruta["color"] =="": #Y para el color
                fruta["color"] = "-desconocido-"

        if fruta["pepas"]:
            pepas = "Presenta pepas en su interior."

        else:
            pepas = "No presenta pepas en su interior."

        return pronombre + " " + fruta["nombre"] + " es una fruta de tipo " + fruta["tipo"] + " de color " + fruta["color"] + ". " + pepas
    
    else: 
        return "Tus variables no estan en el formato requerido, por favor corrigelo"
````

````python
assert generar_descripcion({"nombre":"", "color":"","tipo":"","pepas":True}) != None 
assert generar_descripcion({"nombre":1, "color":True,"tipo":"agua","pepas":"jeje no c"}) != None 

assert generar_descripcion(limon) == "El limón es una fruta de tipo cítrica de color amarillo. Presenta pepas en su interior."
assert generar_descripcion(naranja) == "La naranja es una fruta de tipo cítrica de color naranjo. Presenta pepas en su interior."
assert generar_descripcion(platano) == "El plátano es una fruta de tipo tropical de color amarillo. No presenta pepas en su interior."
assert generar_descripcion(piña) == "La piña es una fruta de tipo tropical de color amarillo. No presenta pepas en su interior."
assert generar_descripcion(frutilla) == "La frutilla es una fruta de tipo bosque de color rojo. Presenta pepas en su interior."
assert generar_descripcion(frambuesa) == "La frambuesa es una fruta de tipo bosque de color rojo. Presenta pepas en su interior."
````

**2. La función `describir(datos)` que generalice la función `generar_descripcion` para recibir una lista de frutas y devolver una lista con las descripciones de cada una:**

Por ejemplo, `describir([limon, naranja]])` deberá devolver una lista de la forma `[descripcion_limon, descripcion_naranja]`

````python
def describir(datos):
    """
    Genera una lista de descripciones en base a una lista de frutas.

    Esta función toma una lista de diccionarios que representan frutas y genera descripciones
    individuales para cada fruta utilizando la función generar_descripcion.

    Parameters:
    -----------
    datos : list
        Una lista de diccionarios, donde cada diccionario contiene información sobre una fruta.
        Cada diccionario debe tener las siguientes claves:
        - 'nombre': El nombre de la fruta.
        - 'tipo': El tipo de la fruta.
        - 'color': El color de la fruta.
        - 'pepas': Un valor booleano que indica si la fruta presenta pepas en su interior.

    Returns:
    --------
    list
        Una lista de descripciones generadas para cada fruta en los datos.
    """
    
    descripciones = []
    for fruta in datos:
        descripciones.append(generar_descripcion(fruta))
        
    return descripciones
````

**3. La función `filtrar_por_pepa(datos, tiene_pepas)` que dado un tipo de fruta y un booleano, retorne las frutas que tienen o no tienen pepas (según el valor de `tiene_pepas`).**
    

Por ejemplo, `filtrar_por_pepa(datos, True)` deberá retornar una lista con los diccionarios de: naranja, limon, frutilla y frambuesa.

````python
def filtrar_por_pepa(datos, tiene_pepas):
    """
    Filtra una lista de frutas según la presencia de pepas.

    Esta función toma una lista de diccionarios que representan frutas y filtra las frutas
    según si tienen o no pepas en su interior, basándose en el valor de 'tiene_pepas'.

    Parameters:
    -----------
    datos : list
        Una lista de diccionarios, donde cada diccionario contiene información sobre una fruta.
        Cada diccionario debe tener las siguientes claves:
        - 'nombre': El nombre de la fruta.
        - 'tipo': El tipo de la fruta.
        - 'color': El color de la fruta.
        - 'pepas': Un valor booleano que indica si la fruta presenta pepas en su interior.
    tiene_pepas : bool
        Un valor booleano que indica si se deben filtrar las frutas con pepas (True) o sin pepas (False).

    Returns:
    --------
    list
        Una lista de diccionarios que representan frutas filtradas según la presencia de pepas.
    """
    
    filtrados = []
    for fruta in datos:
        if fruta["pepas"] == tiene_pepas:
            filtrados.append(fruta)

    return filtrados
````

**4. La función `conteo_colores(datos)` que cree un diccionario que haga un conteo los colores de las frutas.**

Para estos datos, la función debería retornar el siguiente diccionario: 
    
```python
{"amarillo": 3, "naranjo": 1, "rojo": 2}
```

````python
def conteo_colores(datos):
    """
    Realiza un conteo de los colores de las frutas en la lista de datos.

    Esta función toma una lista de diccionarios que representan frutas y cuenta cuántas veces
    aparece cada color en las frutas.

    Parameters:
    -----------
    datos : list
        Una lista de diccionarios, donde cada diccionario contiene información sobre una fruta.
        Cada diccionario debe tener las siguientes claves:
        - 'nombre': El nombre de la fruta.
        - 'tipo': El tipo de la fruta.
        - 'color': El color de la fruta.
        - 'pepas': Un valor booleano que indica si la fruta presenta pepas en su interior.

    Returns:
    --------
    dict
        Un diccionario que contiene un conteo de cuántas veces aparece cada color en las frutas.
    """
    
    colores = []
    for fruta in datos:
        colores.append(fruta["color"])

    conteo = {}
    for color in colores:
        conteo[color] = colores.count(color)

    return conteo
````

**5. La función `obtener_tipos` que devuelva una lista con los tipos únicos de fruta usando `set()`**.

Por ejemplo, `obtener_tipos([limon, naranja, platano])` debería devolver `["citrica", "tropical"]`

````python
def obtener_tipos(datos):
    """
    Obtiene los tipos únicos de frutas presentes en la lista de datos.

    Esta función toma una lista de diccionarios que representan frutas y devuelve una lista
    que contiene los tipos únicos de frutas presentes en los datos.

    Parameters:
    -----------
    datos : list
        Una lista de diccionarios, donde cada diccionario contiene información sobre una fruta.
        Cada diccionario debe tener las siguientes claves:
        - 'nombre': El nombre de la fruta.
        - 'tipo': El tipo de la fruta.
        - 'color': El color de la fruta.
        - 'pepas': Un valor booleano que indica si la fruta presenta pepas en su interior.

    Returns:
    --------
    list
        Una lista que contiene los tipos únicos de frutas presentes en los datos.

    """
   
    tipos = set()
    for fruta in datos:
        tipos.add(fruta["tipo"])
        
    return list(tipos)
````

````python
def obtener_descripciones_v2(datos):
    """
    Genera descripciones para una lista de frutas utilizando comprensiones de lista.

    Esta función toma una lista de diccionarios que representan frutas y genera descripciones
    individuales para cada fruta utilizando comprehensions.

    Parameters:
    -----------
    datos : list
        Una lista de diccionarios, donde cada diccionario contiene información sobre una fruta.
        Cada diccionario debe tener las siguientes claves:
        - 'nombre': El nombre de la fruta.
        - 'tipo': El tipo de la fruta.
        - 'color': El color de la fruta.
        - 'pepas': Un valor booleano que indica si la fruta presenta pepas en su interior.

    Returns:
    --------
    list
        Una lista de descripciones generadas para cada fruta en los datos.
    """

    descripciones = [generar_descripcion(fruta) for fruta in datos]
    return descripciones

def filtrar_por_pepa_v2(datos, tiene_pepas):
    """
    Filtra una lista de frutas según la presencia de pepas utilizando comprensiones de lista.

    Esta función toma una lista de diccionarios que representan frutas y filtra las frutas
    según si tienen o no pepas en su interior, basándose en el valor de 'tiene_pepas', utilizando
    comprehensions.

    Parameters:
    -----------
    datos : list
        Una lista de diccionarios, donde cada diccionario contiene información sobre una fruta.
        Cada diccionario debe tener las siguientes claves:
        - 'nombre': El nombre de la fruta.
        - 'tipo': El tipo de la fruta.
        - 'color': El color de la fruta.
        - 'pepas': Un valor booleano que indica si la fruta presenta pepas en su interior.
    tiene_pepas : bool
        Un valor booleano que indica si se deben filtrar las frutas con pepas (True) o sin pepas (False).

    Returns:
    --------
    list
        Una lista de diccionarios que representan frutas filtradas según la presencia de pepas.
    """

    filtrados = [fruta for fruta in datos if fruta["pepas"] == tiene_pepas]
    return filtrados

def obtener_tipos_v2(datos):
    """
    Obtiene los tipos únicos de frutas presentes en la lista de datos utilizando comprensiones de conjunto.

    Esta función toma una lista de diccionarios que representan frutas y devuelve un conjunto
    que contiene los tipos únicos de frutas presentes en los datos, utilizando comprehensions.

    Parameters:
    -----------
    datos : list
        Una lista de diccionarios, donde cada diccionario contiene información sobre una fruta.
        Cada diccionario debe tener las siguientes claves:
        - 'nombre': El nombre de la fruta.
        - 'tipo': El tipo de la fruta.
        - 'color': El color de la fruta.
        - 'pepas': Un valor booleano que indica si la fruta presenta pepas en su interior.

    Returns:
    --------
    list
        Una lista que contiene los tipos únicos de frutas presentes en los datos.
    """
    
    tipos = [fruta["tipo"] for fruta in datos]
    return tipos
````

````python
class Jugera(Electrodomestico):
    
    def __init__(self):
        super().__init__()
        # Código Aquí
        self.bandeja = []
        
    def agregar_ingrediente(self, nueva_fruta):
        self.bandeja.append(nueva_fruta)
     
    def listar_ingredientes(self):

        nombres_frutas = [fruta["nombre"] for fruta in self.bandeja]

        if len(self.bandeja) != 0:
                    ingredientes_str = ", ".join(nombres_frutas)
                    print(f"Ingredientes en la bandeja: {ingredientes_str}.")
        else:
            print("Bandeja vacía")
    
    def preparar_jugo(self):
        self.esta_enchufado()
    
        if not self.bandeja:
                raise Exception('Error ❌: La bandeja no tiene ingredientes.')

        for fruta in self.bandeja:
            if fruta["pepas"]:
                print('Alerta ⚠️: El jugo puede contener restos de pepas.')
                break
        
        nombres_frutas = [fruta["nombre"] for fruta in self.bandeja]
        ingredientes_str = ", ".join(nombres_frutas)
        
        mensaje = f"Jugo de {ingredientes_str} listo. 🏖️🥤 Que lo disfrutes!!! 🥤🏖️."
        
        self.bandeja = []  # Vaciar la bandeja después de preparar el jugo
        

        return mensaje
````

### Clase Jugera + Properties  [1.0 puntos]

Implementar `bandeja` usando una `property` que permita setear una `nueva_bandeja` como `bandeja` según las siguientes condiciones:


- Compruebe que `nueva_bandeja` sea una lista. En caso contrario, levante una excepción.
- No permita agregar más de 3 ingredientes a la bandeja a la vez. Si se entregan más de 3 frutas, se levante una excepción.
- Se compruebe que todos los elementos de la lista sean frutas. Para esto, por cada fruta compruebe que:
    1. La fruta sea diccionario.
    2. El diccionario entregado tenga las llaves `nombre`, `tipo`, `color` y `pepas`.

````python
class Jugera(Electrodomestico):
    
    def __init__(self):
        super().__init__()
        # Código Aquí
        self.bandeja = []

    @property
    def bandeja(self):
        return self._bandeja
        
    @bandeja.setter
    def bandeja(self, nueva_bandeja):
        if not isinstance(nueva_bandeja, list):
            raise ValueError("La nueva bandeja debe ser una lista.")
        
        if len(nueva_bandeja) > 3:
            raise ValueError("No se pueden agregar más de 3 ingredientes a la bandeja a la vez.")
        
        for fruta in nueva_bandeja:
            if not isinstance(fruta, dict):
                raise ValueError(f"Cada elemento en la lista debe ser un diccionario. Elemento: {fruta} no es un diccionario")
            
            required_keys = {"nombre", "tipo", "color", "pepas"}
            if not all(key in fruta for key in required_keys):
                raise ValueError("Cada diccionario debe tener las llaves 'nombre', 'tipo', 'color' y 'pepas'.")
        
        self._bandeja = nueva_bandeja
        
    def agregar_ingrediente(self, nueva_fruta):
        self.bandeja.append(nueva_fruta)
     
    def listar_ingredientes(self):

        nombres_frutas = [fruta["nombre"] for fruta in self.bandeja]

        if len(self.bandeja) != 0:
                    ingredientes_str = ", ".join(nombres_frutas)
                    print(f"Ingredientes en la bandeja: {ingredientes_str}.")
        else:
            print("Bandeja vacía")
    
    def preparar_jugo(self):
        self.esta_enchufado()
    
        if not self.bandeja:
                raise Exception('Error ❌: La bandeja no tiene ingredientes.')

        for fruta in self.bandeja:
            if fruta["pepas"]:
                print('Alerta ⚠️: El jugo puede contener restos de pepas.')
                break
        
        nombres_frutas = [fruta["nombre"] for fruta in self.bandeja]
        ingredientes_str = ", ".join(nombres_frutas)
        
        mensaje = f"Jugo de {ingredientes_str} listo. 🏖️🥤 Que lo disfrutes!!! 🥤🏖️."
        
        self.bandeja = []  # Vaciar la bandeja después de preparar el jugo
        

        return mensaje
````
